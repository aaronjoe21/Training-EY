# Read JSON and add a column:
## fare_per_km = trip_fare / distance_km
## Write back to JSON

```python
from pyspark.sql.functions import col
df_rides = spark.read.option("header", True).option("inferSchema", True).csv("rides_dataset_csv")
df_rides = df_rides.withColumn("fare_per_km", col("trip_fare") / col("distance_km"))
df_rides.show(5)
df_rides.write.mode("overwrite").json("rides_dataset_with_fare_per_km_json")
```

#Filter JSON data:

```python
df_rides.filter((df_rides.payment_mode=='Card') & (df_rides.vehicle_type=="SUV")).show()
```

## Force JSON output into a single partition and observe the output structure.

```python
from pyspark.sql.functions import col
df_rides = df_rides.withColumn("fare_per_km", col("trip_fare") / col("distance_km"))
df_rides.coalesce(1).write.mode("overwrite").json("rides_dataset_single_json")
```

#Which city generates the highest total trip_fare?

```python
from pyspark.sql.functions import sum
city_revenue = (
    df.groupBy("city")
      .agg(sum("trip_fare").alias("total_trip_fare"))
      .orderBy("total_trip_fare", ascending=False)
)

city_revenue.show(1)
```

# Which vehicle_type has the highest average fare?

```python
from pyspark.sql.functions import avg
avg_fare_highest = (
    df.groupBy("vehicle_type")
      .agg(avg("trip_fare").alias("avg_fare"))
      .orderBy("avg_fare", ascending=False)
)

avg_fare_highest.show(1)
```

## Which driver has completed the most trips?

```python
from pyspark.sql.functions import count, col
completed_trips = df.filter(col("trip_status") == "Completed")

driver_completed = (
    completed_trips.groupBy("driver_name")
                   .agg(count("trip_id").alias("completed_trips"))
                   .orderBy("completed_trips", ascending=False)
)

driver_completed.show(1)
```

# Why is Parquet preferred for analytics dashboards and aggregations?

Parquet is preferred for analytics dashboards and aggregations because it is a columnar, compressed format that reduces storage size, speeds up queries, and minimizes costs by scanning only the necessary columns.

# Repartition the dataset into 4 partitions and write to Parquet

```python

df_repartitioned = df.repartition(4)

df_repartitioned.write.mode("overwrite").parquet("rides_dataset_parquet_4parts")
```

#Create a summary dataset

```python
from pyspark.sql.functions import count, sum, avg, col
df = df.withColumn("trip_fare", col("trip_fare").cast("double")) \
       .withColumn("trip_duration_minutes", col("trip_duration_minutes").cast("double"))

summary_df = (
    df.groupBy("city")
      .agg(
          count("trip_id").alias("total_trips"),
          sum("trip_fare").alias("total_revenue"),
          avg("trip_duration_minutes").alias("average_trip_duration")
      )
)

summary_df.show()

summary_df.write.mode("overwrite").parquet("rides_summary_parquet")
```
