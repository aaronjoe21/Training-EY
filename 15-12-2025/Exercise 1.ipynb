#Add a column:
##delivery_category
#Logic:
###delivery_time_minutes > 45 → "Late"
###else → "OnTime"
##Write output to JSON.

```python
from pyspark.sql.functions import when

df_with_category = df_json.withColumn(
    "delivery_category",
    when(df_json.delivery_time_minutes > 45, "Late").otherwise("OnTime")
)

df_with_category.show()

df_with_category.write.mode("overwrite").json("delivered_orders_with_category_json")
```

# Compare sizes

```python
import os


def get_folder_size(path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return total_size

csv_size = get_folder_size("orders_csv")
json_size = get_folder_size("employees.json")
parquet_size = get_folder_size("employee.parquet")

print("CSV size:", csv_size, "bytes")
print("JSON size:", json_size, "bytes")
print("Parquet size:", parquet_size, "bytes")
```

# Which cuisine generates the highest order_amount overall?

```python
df.write.csv("trailcsv", header=True)
df2 = spark.read.csv("trailcsv", header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import sum

result = (
    df2.groupBy("cuisine")
      .agg(sum("order_amount").alias("total_order_amount"))
      .orderBy("total_order_amount", ascending=False)
)

result.show(1)
```

# Which city has the highest number of orders?

```python
from pyspark.sql.functions import count

city_orders = (
    df.groupBy("city")
      .agg(count("order_id").alias("order_count"))
      .orderBy("order_count", ascending=False)
)

city_orders.show(1)
```

# Which payment mode is most frequently used?

```python
from pyspark.sql.functions import count

payment_mode_counts = (
    df.groupBy("payment_mode")
      .agg(count("order_id").alias("order_count"))
      .orderBy("order_count", ascending=False)
)

payment_mode_counts.show(1)
```

# Why is Parquet the preferred format for analytics platforms like Databricks and BigQuery?

Parquet is preferred because it delivers smaller storage size, faster queries, and efficient compression compared to row‑based formats like CSV or JSON. Analytics platforms such as Databricks and BigQuery rely on these advantages to handle massive datasets cost‑effectively

#Create a report dataset containing:
##city
##total_orders
##total_revenue
#Write it to Parquet.

```python
from pyspark.sql.functions import count, sum

report_df = (
    df2.groupBy("city")
      .agg(
          count("order_id").alias("total_orders"),
          sum("order_amount").alias("total_revenue")
      )
)

report_df.show()
report_df.write.mode("overwrite").parquet("city_report_parquet")
```
